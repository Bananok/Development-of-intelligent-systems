```python
analyzer = {}
for i in range(1, df.shape[0]+1):
  analyzer[i] = []
  for j in df.columns:
    analyzer[i].append(df[j][i])
analyzer
```

```python
def runThrough(stack, word, n):
  for i in range(len(word)):
    
    symbol = word[i]
    print("Now  = ", symbol)

    accept = False
    error = False
    last_symbol = False

    while not(accept) or not(error) or not(last_symbol) :
      print(n, " -> ")

      last_symbol = (symbol == "end" and (symbol in analyzer[n][0])
       and len(stack) == 0)

      symbol = ("const" if symbol.isdigit() else symbol)

      m = ", ) ( true false * / + - or and no end const name".split()
      if i != len(word)-1 and (symbol not in m):
        symbol = ("name" if ((word[i+1] == '(')) else symbol)
      
      symbol = ("variable" if symbol not in m else symbol)
      
      if ((symbol in analyzer[n][0]) and not(last_symbol)):

        new_n = n                  
        if analyzer[n][2] ==  1:
          print(n, "Take from stack")
          new_n = stack.pop()
          print(stack)
        else:
          new_n = analyzer[n][1]

        if analyzer[n][3] ==  1: 
          print(n, "Put in stack")
          stack.append(n+1)
          print(stack)

        if analyzer[n][4] ==  1:   
          accept = True
          print(n, "Accept")
          n = new_n
          break

        n = new_n

      elif last_symbol:
        break
      
      elif analyzer[n][5] == 0:
          n += 1
      else: 
        error = True;
        break
    
    if error: 
      print("This word ∉ formal language")
      break

    if last_symbol:
      print("This word ∈ formal language")
      break
```

```python
def runAnalyzer(word):
  n = 1
  word = word.split()
  word.append("end")
  staсk = deque()
  runThrough(staсk, word, n)

```
